## Explore CitiBike data

### Load CitiBike data

```python
df_bike_monthly=DataFrame.from_csv(os.getenv("PUIDATA") + '/Data/CitiBike_monthly_byStation.csv')
df_bike_monthly.head()
```

### Time Series Aanalysis

```python
plt.rcParams['font.size']=14
f, ax = plt.subplots(figsize=(14,4))
#create a copy of the df for the viz purposes
df_bike_monthly_=df_bike_monthly.copy()
df_bike_monthly_.index=df_bike_monthly_.index.strftime("%B %Y")
df_bike_monthly_.T.boxplot(figsize=(15,7),ax=ax,rot=90, return_type='axes');
ax.set_ylim([-200,16000])
ax.set_title('Fig3 - Monthly Citibike Rides')
```

**Figure 3**: The box plot shows the quartiles of the monthly Citibike ride counts by station. For each of these months, the median rides of the stations are below 2000.

#### Seasonal Decomposition

```python
import statsmodels.api as sm

#Create a monthly time series of bike ride (Average of all citibike stations)
df_new=df_bike_monthly.mean(1).resample('M').mean()
df_new.interpolate(inplace=True)

#Seasonal Decomposition using statsmodels api
res = sm.tsa.seasonal_decompose(df_new.dropna())
plt.rcParams['figure.figsize']=(12,8)
resplot = res.plot()
print ('                                   \
Fig4 - Seasonal Decomposition of Monthly Rides Data')
```

**Figure 4**: The four subplots demonstrate the observed ridership data, trend of the data, seasonal fluctuation, and residual changes of the data with respect to time. It can be seen clearly that there is a cyclical seasonal pattern of the rides.

## Map CitiBike Data to CT level

### Load CitiBike station location data

```python
station_id=DataFrame.from_csv(os.getenv('PUIDATA') + '/Data/Station_id.csv')
#convert station_ids to string index
station_id.index=station_id.index.astype(str)
station_id.head()
```

### Create CitiBike station Geometry (Point) from Corrdinates (Lat & Lon)

```python
import shapely
# combine lat and lon to one column
station_id['lonlat']=zip(station_id.lon,station_id.lat)
# Create Point Geometry for based on lonlat column
station_id['geometry']=station_id[['lonlat']].applymap(lambda x:shapely.geometry.Point(x))
station_id.head()
```

### Intersect the two geometries

```python
a=[]
# check for each citibike stations to see what CT they are belong to
for i in range(len(station_id)):
    a.append(ct_shape[ct_shape.geometry.intersects(station_id.iloc[i].geometry)].BoroCT2010.values[0])
    print '\r',"%",str((i+1)*100./len(station_id))[:4],
station_id['BoroCT2010']=a
```

```python
station_id.BoroCT2010.head()
```

### Add Monthly rides to the station GeoDataFrame with CT code

```python
### Create average monthly rides column from monthly citibike data and add it to the station id which has the geometry
station_id['MonthlyAve']=df_bike_monthly.sum()
```

```python
station_id.head()
```

### Add the monthly average rides to the CT shapefile 

```python
# Create a dataframe "temp" with monthly rides with corresponding station id and borough id
temp=station_id[['MonthlyAve']]
temp.reset_index(inplace=True)
temp.rename(columns = {'index':'Station_id'}, inplace = True)
temp.index = station_id['BoroCT2010']
temp.head(3)
```

```python
# join temp and ct_shape
ct_shape_MonthlyAve=ct_shape.join(temp)
ct_shape_MonthlyAve.fillna(0,inplace=True)
ct_shape_MonthlyAve.head(2)
```

```python
f, ax = plt.subplots(figsize=(8,8))
ct_shape_MonthlyAve.plot(column='MonthlyAve',colormap='hot_r',alpha=1,linewidth=0.1,ax=ax)
plt.title("Fig5 - Monthly Ride Counts")
```

## Exploratory Geospatial Analysis

```python
import pysal as ps
```

### Spatial Weights
Contiguity: Queen Weights
-- Queen contiguity defines as neighbors any pair of polygons that share at
least one vertex in their polygon definitions

```python
qW_CT=ps.queen_from_shapefile(path)
```

#### Neighbors of index10

```python
qW_CT[10]
```

```python
qW_CT.cardinalities[10]
```

#### Matrix of Weights

```python
Wmatrix, ids = qW_CT.full()
```

```python
Wmatrix
```

#### Change the matrix to be row-standardized

```python
qW_CT.transform = 'r'
```

### Spatial Lag  
Similarity of CT neibors can be meseard by means of spatial lag as follow:
$$BikeRideCountLag = \sum_j w_{i,j} BikeRideCount$$

```python
## Calculate the spatial lag by adding neiboughers's weights to the monthly rides
MonthlyRide=ct_shape_MonthlyAve_Zoom.MonthlyAve
MonthlyRideLag=ps.lag_spatial(qW_CT, MonthlyRide)
```

#### Create Quantiles for maping

```python
MonthlyRideLagQ10 = ps.Quantiles(MonthlyRideLag, k=10)
MonthlyRideLagQ10
```

#### Map the monthly rides lag

```python
f, ax = plt.subplots(figsize=(9, 9))
ct_shape_MonthlyAve_Zoom.assign(cl=MonthlyRideLagQ10.yb).plot(column='cl', categorical=True, 
        k=10, cmap='hot_r', linewidth=0.1, ax=ax, 
        edgecolor='white', legend=True)
ax.set_axis_off()
plt.title("Fig7 - Monthly Ride Spatial Lag Deciles")
```

**Figure 7**: The map shows the study area that is colored based on 10 quantiles of **monthly** ride spatial lag. Darker color indicates larger rideships of the spatial lag quantile.

## Global  Spatial Autocorrelation

```python
b,a = np.polyfit(MonthlyRide, MonthlyRideLag, 1)
b,a 
```

### Moran Scatterplot

```python
f, ax = plt.subplots(1, figsize=(9, 9))

plt.plot(MonthlyRide, MonthlyRideLag, '.', color='firebrick')

 # dashed vert at mean of the last year's PCI
plt.vlines(MonthlyRide.mean(), MonthlyRideLag.min(), MonthlyRideLag.max(), linestyle='--')
 # dashed horizontal at mean of lagged PCI
plt.hlines(MonthlyRideLag.mean(), MonthlyRide.min(), MonthlyRide.max(), linestyle='--')

# red line of best fit using global I as slope
plt.plot(MonthlyRide, a + b*MonthlyRide, 'r')
plt.title('Fig8 - Moran Scatterplot')
plt.ylabel('Spatial Lag of MonthlyRides')
plt.xlabel('MonthlyRides')
```

**Figure 8**: The map shows the correlation between monthly rides and monthly rides spatial lag with a fitted line.
The dotted PCI lines divide the areas into four where the top right area contains the **hot spot** data points and bottom left are **cold spot** data points.

### Moran's I

```python
I_MonthlyRide = ps.Moran(MonthlyRide, qW_CT)
```

```python
I_MonthlyRide.I, I_MonthlyRide.p_sim
```

Thus, the $I$ statistic is $0.656$ for this data, and has a very small $p$ value. 

```python
I_MonthlyRide.sim[0:10]
```

### KDEplot of all of the simulated points
the vertical line shows the observed value

```python
import seaborn as sns
plt.rcParams['figure.figsize']=(14,3)
```

```python
sns.kdeplot(I_MonthlyRide.sim, shade=True)
plt.vlines(I_MonthlyRide.sim, 0, 1)
plt.vlines(I_MonthlyRide.I, 0, 10, 'r')
plt.xlim([-.70, .70])
plt.title('Fig9-1 - KDE Plot for Simulated Monthly Data Points w. Observed MI')
```

**Figure 9-1**: Bell shaped simulated data points of monthly rides versus the red vertical line--the observed value of Moran's I. (See more explanation below)

In the following figure the vertical line is showing if our $I$ statistic were close to our expected value

```python
sns.kdeplot(I_MonthlyRide.sim, shade=True)
plt.vlines(I_MonthlyRide.sim, 0, 1)
plt.vlines(I_MonthlyRide.EI+.01, 0, 10, 'r')
plt.xlim([-.70, .70])
plt.title('Fig9-2 - KDE Plot for Simulated Monthly Data Points w. Expected MI')
```

**Figure 9-2**: Bell shaped simulated data points of monthly rides with the expected value of Moran's I under normality assumption(the red vertical line). 

The observed value for I is significantly above its expected value, under the assumption of normality for the monthly rides.

## Local  Spatial Autocorrelation

```python
ps.version #Check verison. The version needs to be 1.12 to correctly run ps.Moran_Local
```

```python
LMo_MonthlyRide = ps.Moran_Local(MonthlyRide, qW_CT, permutations=999)
```

```python
# p_sim: p-values based on permutations (one-sided)
LMo_MonthlyRide.Is[0:5], LMo_MonthlyRide.p_sim[0:5]
```

### LISA (Local Indicators of Spatial Autocorrelation)

```python
# Classify significant and non-significant rides based on value of Local Moran
sigs = MonthlyRide[LMo_MonthlyRide.p_sim <= .001]
W_sigs = MonthlyRideLag[LMo_MonthlyRide.p_sim <= .001]
insigs = MonthlyRide[LMo_MonthlyRide.p_sim > .001]
W_insigs = MonthlyRideLag[LMo_MonthlyRide.p_sim > .001]
```

### Plot Signficant LISA

```python
f, ax = plt.subplots(figsize=(9, 9))
plt.plot(sigs, W_sigs, '.', color='firebrick')
plt.plot(insigs, W_insigs, '.k', alpha=.2)
plt.text(s='$I = %.3f$' % I_MonthlyRide.I, x=120000, y=5000, fontsize=18)

 # dashed vert at mean of the last year's PCI
plt.vlines(MonthlyRide.mean(), MonthlyRideLag.min(), MonthlyRideLag.max(), linestyle='--')
 # dashed horizontal at mean of lagged PCI
plt.hlines(MonthlyRideLag.mean(), MonthlyRide.min(), MonthlyRide.max(), linestyle='--')

# red line of best fit using global I as slope
plt.plot(MonthlyRide, a + b*MonthlyRide, 'r')
plt.title('Fig10 - Moran Scatterplot (Significant points colored red)')
plt.ylabel('Spatial Lag of MonthlyRides')
plt.xlabel('MonthlyRides')
```

**Figure 10**: The plot demonstrates the spatial lag of monthly rides versus the monthly rides data points with a fitted line. The significant rides as colored as red while the insignificant ones are colred as light grey based on Local Moran's p-values.

### Hot and Cold spots 

```python
sig = LMo_MonthlyRide.p_sim < 0.01
hotspots = LMo_MonthlyRide.q==1 * sig
coldspots = LMo_MonthlyRide.q==3 * sig
print "Number of Hotspots:", hotspots.sum()
print "Number of Coldspots:", coldspots.sum()
```

```python
from matplotlib import colors

hcmap = colors.ListedColormap(['grey','red','blue'])
hotcold = hotspots * 1 + coldspots * 2

f, ax = plt.subplots(figsize=(9, 9))
ct_shape_MonthlyAve_Zoom.assign(cl=hotcold).plot(column='cl', categorical=True, 
        k=10, cmap=hcmap, linewidth=0.1, ax=ax, 
        edgecolor='white', legend=True)
ax.set_axis_off()
plt.title("Fig11 - Monthly Average Ride Spatial Lag Deciles")
```

**Figure 11**: The heat maps show the hot spots and cold spots of Citi-bike monthly rides based on Moran's I Statistics(Local), indicated by the red and blue colored regions. Auto-correlation occurs between the ridership within a borough tract and its neighbour tracts(neighbours are weighted). The cluster of red regions indicates that the tracts with high bike ridership are likely to have adjacent tracts also having high riderships; the cluster of blue regions shows that the tracts with low ridership are more likely to have low ridership neighbour tracts.
